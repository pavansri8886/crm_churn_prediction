# 📖 PARTIE 1 — Fondements (Chapitre 1)
## Introduction au Data Mining et à la prise de décision

### 1.1 Pourquoi le Data Mining ?

Les entreprises de télécommunications génèrent des **volumes massifs de données** chaque jour :
- Historique d'appels, SMS, consommation internet
- Facturation mensuelle, historique de paiements
- Informations contractuelles (type de contrat, ancienneté)

Seules, ces données brutes n'ont pas de valeur. Le **Data Mining** permet de les transformer en **connaissance actionnable** :

```
Données brutes → Information → Connaissance → Décision
```

Dans notre cas :  
- **Données** : 7 043 clients, 20 variables (contrat, charges, services...)
- **Information** : les clients en contrat mensuel paient en moyenne 65€/mois
- **Connaissance** : les clients en contrat mensuel + fibre optique + sans sécurité ont 45% de probabilité de churn
- **Décision** : lancer une campagne de rétention ciblée sur ce segment

### 1.2 Positionnement dans l'écosystème

```
Big Data
  └── Business Intelligence (reporting, dashboards)
        └── Data Mining (découverte de patterns)
              └── Machine Learning (modèles prédictifs)
                    └── Data Science (approche scientifique complète)
```

Notre projet couvre **toute cette chaîne** : de la donnée brute jusqu'au modèle déployé en production.

# 🔵 PHASE 1 — Business Understanding

## Compréhension métier

### Contexte
Une entreprise de télécommunications perd des clients chaque mois (**churn**). Chaque client perdu représente un manque à gagner direct. Il est beaucoup moins coûteux de **retenir un client existant** que d'en acquérir un nouveau.

### Objectif métier
> Identifier, **en avance**, les clients susceptibles de résilier leur abonnement afin de déclencher des actions de rétention ciblées (offre commerciale, appel client, réduction tarifaire).

### Traduction en problème Data Mining
| Dimension | Détail |
|-----------|--------|
| **Type de problème** | Classification supervisée (binaire) |
| **Variable cible** | Churn : Yes (1) / No (0) |
| **Critère de succès** | ROC-AUC > 0.75, F1-score > 0.55 |
| **Usage du modèle** | Score de risque mensuel par client |
| **Action déclenchée** | Campagne rétention si probabilité > 35% |

### Hypothèse métier
Comme dans l'exemple CRISP-DM du cours (factures impayées), nous posons l'hypothèse que **les clients qui vont churner changent leur comportement** avant de partir :
- Contrat court terme + services minimaux = signal de départ imminent
- Charges élevées sans engagement long terme = client mécontent
- Nouveau client sans add-ons = risque d'attrition fort

# 🔵 PHASE 3 — Data Preparation

## Préparation des données

Cette phase correspond directement au **Step 2 : Scrub the Dirt** du cours de Data Cleaning :
- Missing Data → Imputation
- Data Type Issues → Coercion de type
- Inconsistent Data → Encodage uniforme
- Feature Engineering → Création de variables métier

````python
# ── 3.1 Nettoyage des données ─────────────────────────────────────────────────
df_clean = df.copy()

# Suppression de l'identifiant client (aucune valeur prédictive)
df_clean.drop(columns=['customerID'], inplace=True)

# Correction du type de TotalCharges (string → float)
df_clean['TotalCharges'] = pd.to_numeric(df_clean['TotalCharges'], errors='coerce')

# Imputation des valeurs manquantes
df_clean['TotalCharges'].fillna(df_clean['TotalCharges'].median(), inplace=True)

# Encodage de la variable cible
df_clean['Churn'] = df_clean['Churn'].map({'Yes': 1, 'No': 0})

print('✅ Nettoyage terminé')
print(f'Valeurs manquantes restantes : {df_clean.isnull().sum().sum()}')
print(f'Shape : {df_clean.shape}')
````

````python
# ── 3.2 Feature Engineering — Variables métier CRM ───────────────────────────
# Création de 7 nouvelles variables à partir de la connaissance métier

# 1. Coût moyen par mois d'ancienneté
df_clean['ChargesPerTenureMonth'] = df_clean['MonthlyCharges'] / (df_clean['tenure'] + 1)

# 2. Indicateur nouveau client (< 6 mois)
df_clean['IsNewCustomer'] = (df_clean['tenure'] <= 6).astype(int)

# 3. Indicateur client fidèle (> 24 mois)
df_clean['IsLongTerm'] = (df_clean['tenure'] >= 24).astype(int)

# 4. Score de bundle de services (nombre de services actifs)
service_cols = ['PhoneService', 'OnlineSecurity', 'OnlineBackup',
                'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']
df_clean['ServiceBundleScore'] = df_clean[service_cols].apply(
    lambda x: (x == 'Yes').sum(), axis=1
)

# 5. Score de risque contractuel
contract_risk = {'Month-to-month': 3, 'One year': 2, 'Two year': 1}
df_clean['ContractRiskScore'] = df_clean['Contract'].map(contract_risk)

# 6. Écart entre charges totales attendues et réelles
expected_total = df_clean['MonthlyCharges'] * df_clean['tenure']
df_clean['TotalChargesGap'] = expected_total - df_clean['TotalCharges']

# 7. Combinaison haute risque : contrat mensuel + fibre + pas de sécurité
df_clean['HighRiskCombo'] = (
    (df_clean['Contract'] == 'Month-to-month') &
    (df_clean['InternetService'] == 'Fiber optic') &
    (df_clean['OnlineSecurity'] == 'No')
).astype(int)

print('✅ Feature Engineering terminé — 7 nouvelles variables créées :')
new_features = ['ChargesPerTenureMonth', 'IsNewCustomer', 'IsLongTerm',
                'ServiceBundleScore', 'ContractRiskScore', 'TotalChargesGap', 'HighRiskCombo']
for f in new_features:
    print(f'   • {f}')
````

````python
# ── 3.3 Encodage et préparation du pipeline sklearn ──────────────────────────
X = df_clean.drop(columns=['Churn'])
y = df_clean['Churn']

# Identification des colonnes numériques et catégorielles
numerical_cols   = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()

print(f'Variables numériques ({len(numerical_cols)}) : {numerical_cols}')
print(f'Variables catégorielles ({len(categorical_cols)}) : {categorical_cols}')

# Pipeline de prétraitement
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_cols)
])
````

````python
# ── 3.4 Split stratifié 70 / 10 / 20 ─────────────────────────────────────────
X_train_val, X_test, y_train_val, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)
X_train, X_val, y_train, y_val = train_test_split(
    X_train_val, y_train_val, test_size=0.125, random_state=42, stratify=y_train_val
)

print('✅ Données séparées (stratifié — même taux de churn dans chaque split) :')
print(f'   Train      : {len(X_train):,} clients ({len(X_train)/len(X)*100:.0f}%) — churn rate: {y_train.mean():.1%}')
print(f'   Validation : {len(X_val):,} clients ({len(X_val)/len(X)*100:.0f}%) — churn rate: {y_val.mean():.1%}')
print(f'   Test       : {len(X_test):,} clients ({len(X_test)/len(X)*100:.0f}%) — churn rate: {y_test.mean():.1%}')
````

---
# 🔵 PHASE 4 — Modeling

## Modélisation ML

Conformément au cours (Chapitre 2 — Types de Machine Learning), notre problème est un cas d'**apprentissage supervisé de classification** :
- Les données sont **labellisées** (Churn = Yes/No connu)
- L'objectif est de **prédire une catégorie** pour de nouveaux clients

Nous entraînons 4 modèles pour comparaison :

| Modèle | Type | Caractéristique |
|--------|------|----------------|
| **Logistic Regression** | Linéaire | Simple, interprétable |
| **Random Forest** | Ensemble (bagging) | Robuste, gère les non-linéarités |
| **XGBoost** | Ensemble (boosting) | Haute performance |
| **LightGBM** | Ensemble (boosting) | Rapide sur grands datasets |

````python
# ── 4.1 Définition des modèles ────────────────────────────────────────────────
models = {
    'Logistic Regression': Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', LogisticRegression(max_iter=1000, random_state=42))
    ]),
    'Random Forest': Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1))
    ]),
    'XGBoost': Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', XGBClassifier(
            n_estimators=100, random_state=42,
            eval_metric='logloss', verbosity=0
        ))
    ]),
    'LightGBM': Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', LGBMClassifier(n_estimators=100, random_state=42, verbose=-1))
    ]),
}

print(f'✅ {len(models)} modèles définis')
````

````python
# ── 4.2 Entraînement et évaluation ───────────────────────────────────────────
results = {}
trained_models = {}

for name, pipeline in models.items():
    print(f'🔄 Entraînement : {name}...')
    pipeline.fit(X_train, y_train)
    
    # Prédictions sur validation
    y_val_proba = pipeline.predict_proba(X_val)[:, 1]
    y_val_pred  = (y_val_proba >= 0.5).astype(int)
    
    # Prédictions sur test
    y_test_proba = pipeline.predict_proba(X_test)[:, 1]
    y_test_pred  = (y_test_proba >= 0.5).astype(int)
    
    results[name] = {
        'Val ROC-AUC' : round(roc_auc_score(y_val,  y_val_proba),  4),
        'Test ROC-AUC': round(roc_auc_score(y_test, y_test_proba), 4),
        'Test F1'     : round(f1_score(y_test,  y_test_pred),  4),
        'Test Precision': round(precision_score(y_test, y_test_pred), 4),
        'Test Recall' : round(recall_score(y_test, y_test_pred), 4),
    }
    trained_models[name] = pipeline
    print(f'   ✅ Test ROC-AUC = {results[name]["Test ROC-AUC"]} | F1 = {results[name]["Test F1"]}')

results_df = pd.DataFrame(results).T
print('\n' + '='*60)
print('TABLEAU COMPARATIF DES MODÈLES')
print('='*60)
print(results_df.to_string())
````

---
# 🔵 PHASE 5 — Evaluation

## Évaluation et sélection du meilleur modèle

````python
# ── 5.1 Courbes ROC — tous les modèles ───────────────────────────────────────
plt.figure(figsize=(10, 7))

for (name, pipeline), color in zip(trained_models.items(), COLORS):
    y_proba = pipeline.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    auc = roc_auc_score(y_test, y_proba)
    plt.plot(fpr, tpr, color=color, lw=2, label=f'{name} (AUC = {auc:.3f})')

plt.plot([0, 1], [0, 1], 'k--', lw=1, label='Modèle aléatoire (AUC = 0.5)')
plt.fill_between([0,1],[0,1],[1,1], alpha=0.05, color='green')
plt.xlabel('Taux de Faux Positifs (FPR)', fontsize=12)
plt.ylabel('Taux de Vrais Positifs (TPR)', fontsize=12)
plt.title('Courbes ROC — Comparaison des 4 modèles\n(Zone verte = meilleure performance)', 
          fontsize=13, fontweight='bold')
plt.legend(loc='lower right', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('roc_curves.png', dpi=150, bbox_inches='tight')
plt.show()
````

````python
# ── 5.2 Sélection du meilleur modèle ─────────────────────────────────────────
best_model_name = results_df['Test ROC-AUC'].idxmax()
best_model      = trained_models[best_model_name]

print(f'🏆 Meilleur modèle : {best_model_name}')
print(f'   ROC-AUC  = {results_df.loc[best_model_name, "Test ROC-AUC"]}')
print(f'   F1-Score = {results_df.loc[best_model_name, "Test F1"]}')
print(f'   Precision = {results_df.loc[best_model_name, "Test Precision"]}')
print(f'   Recall   = {results_df.loc[best_model_name, "Test Recall"]}')
````

````python
# ── 5.3 Matrice de confusion ──────────────────────────────────────────────────
y_pred_best = best_model.predict(X_test)
cm = confusion_matrix(y_test, y_pred_best)

plt.figure(figsize=(7, 5))
sns.heatmap(
    cm, annot=True, fmt='d', cmap='Blues',
    xticklabels=['Prédit: No Churn', 'Prédit: Churn'],
    yticklabels=['Réel: No Churn', 'Réel: Churn']
)
plt.title(f'Matrice de confusion — {best_model_name}', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=150, bbox_inches='tight')
plt.show()

tn, fp, fn, tp = cm.ravel()
print(f'\n📊 Interprétation :')
print(f'   Vrais Négatifs  (TN) = {tn:,}  → clients non-churners correctement identifiés')
print(f'   Faux Positifs   (FP) = {fp:,}  → clients fidèles incorrectement ciblés par rétention')
print(f'   Faux Négatifs   (FN) = {fn:,}  → churners manqués (risque financier)')
print(f'   Vrais Positifs  (TP) = {tp:,}  → churners correctement détectés')
````

````python
# ── 5.4 Rapport de classification complet ────────────────────────────────────
print('RAPPORT DE CLASSIFICATION COMPLET')
print('=' * 55)
print(classification_report(y_test, y_pred_best,
                             target_names=['No Churn (0)', 'Churn (1)']))
````

````python
# ── 5.5 Importance des features (SHAP) ───────────────────────────────────────
# Préparation des données transformées
preprocessor_fitted = best_model.named_steps['preprocessor']
X_test_transformed  = preprocessor_fitted.transform(X_test)

# Récupération des noms de features
num_feature_names = numerical_cols
cat_feature_names = preprocessor_fitted.named_transformers_['cat'].get_feature_names_out(categorical_cols).tolist()
all_feature_names = num_feature_names + cat_feature_names

# SHAP
classifier = best_model.named_steps['classifier']

try:
    explainer = shap.TreeExplainer(classifier)
except:
    explainer = shap.LinearExplainer(classifier, X_test_transformed)

shap_values = explainer.shap_values(X_test_transformed[:200])
if isinstance(shap_values, list):
    shap_values = shap_values[1]

plt.figure(figsize=(10, 7))
shap.summary_plot(
    shap_values, X_test_transformed[:200],
    feature_names=all_feature_names,
    max_display=15, show=False
)
plt.title('Importance des variables (SHAP) — Impact sur la prédiction du churn',
          fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig('shap_importance.png', dpi=150, bbox_inches='tight')
plt.show()
print('💡 Les variables en rouge augmentent le risque de churn, celles en bleu le réduisent.')
````

````python
# ── 5.6 Visualisation comparative des scores ──────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

metrics = ['Test ROC-AUC', 'Test F1']
titles  = ['ROC-AUC Score (test)', 'F1-Score (test)']

for ax, metric, title in zip(axes, metrics, titles):
    vals  = results_df[metric]
    bars  = ax.bar(vals.index, vals.values, color=COLORS, edgecolor='white', linewidth=1.5)
    ax.set_title(title, fontsize=13, fontweight='bold')
    ax.set_ylim(0, 1)
    ax.tick_params(axis='x', rotation=15)
    for bar, val in zip(bars, vals.values):
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
                f'{val:.3f}', ha='center', fontweight='bold')
    # Highlight best
    best_idx = list(vals.index).index(vals.idxmax())
    bars[best_idx].set_edgecolor('gold')
    bars[best_idx].set_linewidth(3)

plt.suptitle(f'Comparaison des 4 modèles — Meilleur : {best_model_name} ⭐',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('model_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
````

````python
# ── 6.2 Test de prédiction sur un nouveau client ──────────────────────────────
# Simulation d'un client à haut risque
new_customer = pd.DataFrame([{
    'SeniorCitizen'  : 0,
    'gender'         : 'Male',
    'Partner'        : 'No',
    'Dependents'     : 'No',
    'tenure'         : 2,
    'PhoneService'   : 'Yes',
    'MultipleLines'  : 'No',
    'InternetService': 'Fiber optic',
    'OnlineSecurity' : 'No',
    'OnlineBackup'   : 'No',
    'DeviceProtection':'No',
    'TechSupport'    : 'No',
    'StreamingTV'    : 'No',
    'StreamingMovies': 'No',
    'Contract'       : 'Month-to-month',
    'PaperlessBilling': 'Yes',
    'PaymentMethod'  : 'Electronic check',
    'MonthlyCharges' : 95.0,
    'TotalCharges'   : 190.0,
}])

# Ajout des features engineered
new_customer['ChargesPerTenureMonth'] = new_customer['MonthlyCharges'] / (new_customer['tenure'] + 1)
new_customer['IsNewCustomer']   = (new_customer['tenure'] <= 6).astype(int)
new_customer['IsLongTerm']      = (new_customer['tenure'] >= 24).astype(int)
new_customer['ServiceBundleScore'] = new_customer[service_cols].apply(lambda x: (x=='Yes').sum(), axis=1)
new_customer['ContractRiskScore']  = new_customer['Contract'].map(contract_risk)
new_customer['TotalChargesGap']    = new_customer['MonthlyCharges'] * new_customer['tenure'] - new_customer['TotalCharges']
new_customer['HighRiskCombo'] = (
    (new_customer['Contract'] == 'Month-to-month') &
    (new_customer['InternetService'] == 'Fiber optic') &
    (new_customer['OnlineSecurity'] == 'No')
).astype(int)

churn_proba = best_model.predict_proba(new_customer)[0, 1]

risk_level = 'ÉLEVÉ 🔴' if churn_proba > 0.66 else ('MOYEN 🟡' if churn_proba > 0.33 else 'FAIBLE 🟢')

print('=' * 55)
print('PRÉDICTION POUR UN NOUVEAU CLIENT')
print('=' * 55)
print(f'Probabilité de churn : {churn_proba:.1%}')
print(f'Niveau de risque     : {risk_level}')
print()
if churn_proba > 0.33:
    print('💡 Recommandation CRM : Déclencher une action de rétention')
    print('   → Proposer une offre de fidélisation (remise 15%)')
    print('   → Appel du service client dans les 7 jours')
    print('   → Offre de migration vers contrat 1 an')
else:
    print('✅ Client stable — Aucune action prioritaire requise')
````
